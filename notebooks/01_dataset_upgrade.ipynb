```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_excel(
    "../data/raw/customer_churn.xlsx"
)
df.head()
```

```python
#check for missing values
df.isnull().sum()
```

```python
#check for duplicate customers
df.duplicated().sum()
```

```python
#totalcharges is shown as object. should be number
df["TotalCharges"] = pd.to_numeric(
    df["TotalCharges"],
    errors="coerce"
)
df.info()
```

```python
#fill any missing charges with median. as median is better than man value for skewed data
df['TotalCharges'] = (df['TotalCharges'].fillna(df['TotalCharges'].median()))
df.head()
```

```python
df.isnull().sum()
```

```python
#now standardize all categorical values. fix inconsistencies for cat values like hidden spaces,etc
categorical_vals = df.select_dtypes(include='object').columns
for col in categorical_vals:
    df[col] = df[col].str.strip()
```

```python
#now adding new advanced cols
#CUSTOMER LIFETIME VALUES(CLV) = MonthlyCharges * tenure
df['customer_lifetime_value'] = (df['MonthlyCharges'] * df['tenure'])
df.head()
```

```python
#REVENUE TIER
conditions = [
    df['customer_lifetime_value'] < 1000,

    (df['customer_lifetime_value'] >= 1000) & (df['customer_lifetime_value'] < 3000),

    df['customer_lifetime_value'] > 3000
]

choices = ['Bronze', 'Silver', 'Gold']

df['revenue_tier'] = np.select(conditions,choices,default='Bronze')
```

```python
df.head()
```

```python
#SUPPORT FRUSTRATION SCORE = tech_ticket *2 + admin_ticket
#Tech problems hurt retention more.
df['support_frustration_score'] = (df['numTechTickets'] * 2 + df['numAdminTickets'])
df.head()
```

```python
#SERVICE ADOPTION SCORE = More service usage = lower churn
service_columns = [
    "PhoneService",
    "OnlineSecurity",
    "OnlineBackup",
    "DeviceProtection",
    "TechSupport",
    "StreamingTV",
    "StreamingMovies"
]

#convert yes -> 1
for col in service_columns :
    df[col] = np.where(
        df[col] == 'Yes',
        1,
        0
    )

#create score 
df['service_adoption_score'] = (
    df[service_columns].sum(axis=1)
)

df.head()
```

```python
#BILLING RISK SCORE  = Risk indicators:Month-to-month contract,Electronic check,Paperless billing
#higher score = higher churn
df['billing_risk_score'] = 0

df.loc[
    df['Contract'] == 'Month-to-month',
    'billing_risk_score'
] += 3

df.loc[
    df['PaymentMethod'] == 'Electronic check',
    'billing_risk_score'
] += 2

df.loc[
    df['PaperlessBilling'] == 'Yes',
    'billing_risk_score'
] +=1

```

```python
#ENGAGEMENT SCORE = tenure + service adoption - frustration
#higher engagemnet = stickier customer

df['engagement_score'] = (
    (df['tenure'] * 0.5)
    +
    (df['service_adoption_score'] * 5)
    -
    (df['support_frustration_score'] * 2)
)

df.head()
```

```python
#HIGH VALUE CHURN FLAG = expensive customer we lost

df['high_value_churn_flag'] = np.where(
    (df['revenue_tier'] == 'Gold')
    &
    (df['Churn'] == 'Yes'),
    1,
    0
)

df.head()
```

```python
#CHURN RISK SEGMENT

risk_score = (
    df["billing_risk_score"]
    +
    df["support_frustration_score"]
)


conditions = [
    risk_score <= 3,
    (risk_score > 3)
    & (risk_score <= 6),
    risk_score > 6
]

choices = [
    "Low Risk",
    "Medium Risk",
    "High Risk"
]

df["churn_risk_segment"] = np.select(
    conditions,
    choices,
    default="Unknown"
)

df.head()
```
